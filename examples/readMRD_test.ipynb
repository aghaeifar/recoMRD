```python
# this notebook tests the core script, readMRD, and shows sampling pattern

# if jupyterlab:
%matplotlib widget
# if jupyter notebook:
# %matplotlib notebook
import numpy as np 
from recoMRD import readMRD
import matplotlib.pyplot as plt
from bart import bart
import os
```

```python
folder   = './../example_data/'
mrd_files  = [each for each in os.listdir(folder) if each.endswith('.mrd')]

for filename in mrd_files:
    print(f'Reading {filename}')
    mrd = readMRD(os.path.join(folder, filename))
    for scan in mrd.kspace:
        print(f'Size {scan} = {mrd.kspace[scan].shape}' )
    
    
    temp = mrd.kspace['image_scan'][0,:,:,:,:,0,0,0,0,0,0].squeeze()
    mask_img = bart.bart(1, 'pattern', temp).real

    if mrd.kspace.get('acs') is not None:
        temp = mrd.kspace.get('acs')[0,:,:,:,:,0,0,0,0,0,0].squeeze()
        mask_acs = bart.bart(1, 'pattern', temp).real
        print(f'Img size = {mrd.kspace.get("image_scan").shape}')
        print(f'ACS size = {mrd.kspace.get("acs").shape}')

    #pre, ext = os.path.splitext(filename)
    #img = nib.Nifti1Image(mask.squeeze(), affine=np.eye(4))
    #nib.save(img, os.path.join(folder, pre+".nii"))

    fig, axes = plt.subplots(1,2,figsize=(10, 5))
    axes[0].imshow(np.abs(mask_img[:,:,mask_img.shape[2]//2]).T, vmin=0, vmax=1)
    axes[0].set_title('Image')
    if mrd.kspace.get('acs') is not None:
        axes[1].imshow(np.abs(mask_acs[:,:,mask_acs.shape[2]//2]).T, vmin=0, vmax=1)
        axes[1].set_title('ACS')
    else:
        axes[1].set_visible(False)

    plt.tight_layout()
    plt.show()

    print('----------------------------------------------------------------------------------------------------')
```
